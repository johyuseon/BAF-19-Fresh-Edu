# BAF-19-Fresh-Edu

## data 1
## 데이터명: Inventory Analysis Case Study
## 데이터 특징
    - 출처: kaggle(https://www.kaggle.com/datasets/bhanupratapbiswas/inventory-analysis-case-study)
    - 분야: 재고 관리
    - 내용: 전자 부품을 생산하는 중견 제조업체인 Any Manufacturing Company의 재고 분석을 위한 데이터이다. 기초 재고, 기말 재고, 판매 이력 등 재고에 대한 전반적인 거래 내용이 포함되어 있다. 따라서 과거 데이터를 활용해 재고 판매 수준을 예측할 수 있다. 또한 원자재, WIP, 완제품에 대한 데이터로 재주문 시점과 주문량, 리드 타임 등을 구할 수 있어서 재고 관리 모델에 대한 분석이 가능하다. 이를 토대로 기업의 재고 관리 전략을 세울 수 있다.

## data 2
## 데이터명: 서울시 문화활동 연간 방문횟수 통계
## 데이터 특징
    - 출처: 서울 열린데이터 광장 (https://data.seoul.go.kr/dataList/DT201011H040009/S/2/datasetView.do)
    - 분야: 문화
    - 내용: 이 데이터에는 성별, 연령, 학력, 소득, 지역 등 여러 요인에 따른 문화 활동 연간 방문 횟수를 나타낸다. 문화 활동은 전시회, 박물관, 대중 공연 등을 포함한다. 다양한 요인들을 활용해 어떤 범주의 사람들이 특정 문화 활동을 많이 방문하는지 분석하여 문화 시설의 마케팅에도 활용할 수 있을 것이다.
